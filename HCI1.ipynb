```
import skimage
```

```
from skimage import io
img=io.imread('/ironman.jpg')
io.imshow(img)
img.shape
```

```
from skimage import exposure
from pylab import *
gama1=exposure.adjust_gamma(img,0.5)
gama2=exposure.adjust_gamma(img,5)
figure(0)
io.imshow(gama1)
figure(1)
io.imshow(gama2)
```

```
from skimage.transform import rotate
img_rot=rotate(img,45)
io.imshow(img_rot)
```

```
import pandas as pd
df=pd.DataFrame(img.flatten())
df
```

```
from skimage import data
from skimage import color
img_xyz=color.rgb2xyz(img)
img_rgb=color.xyz2rgb(img_xyz)
figure(0)
io.imshow(img_xyz)
figure(1)
io.imshow(img_rgb)
```

```
#drawing Beziar curve
from skimage import draw
x,y=draw.bezier_curve(0,0,500,500,700,700,100)
img[x,y]=1
io.imshow(img)
```

```
# resize image 
from skimage.transform import resize
image_resized = resize(img, (img.shape[0] // 4, img.shape[1] // 4),
                       anti_aliasing=True)
io.imshow(image_resized)
```

```
#image structural comparison
from skimage.metrics import structural_similarity as ssim
img1=io.imread('ironman1.jpg') #(413, 620, 3)
img2=io.imread('warmachine.jpg') #(828, 576, 3)

figure(0)
io.imshow(img1)
figure(1)
io.imshow(img2)
#since the sizes of both are different lets resize them to same dimensions say 500X500
img1_resized=resize(img1,(500,500),anti_aliasing=True)
img2_resized=resize(img2,(500,500),anti_aliasing=True)

figure(2)
io.imshow(img1_resized)
figure(3)
io.imshow(img2_resized)
score=ssim(img1_resized,img2_resized,multichannel=True)
print(score)
```
