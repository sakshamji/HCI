```
#getting image and converting it to gray scale image
from skimage import exposure,io,color
from pylab import *
img=io.imread('rider.jpg')
figure(0)
io.imshow(img)
gray_img=color.rgb2gray(img)
figure(1)
io.imshow(gray_img)
```

```
#applying gaussian filter
from skimage import filters
from scipy.ndimage import filters
img_g1=filters.gaussian_filter(gray_img,20)
figure(0)
io.imshow(img_g1)
img_g2=filters.gaussian_filter(gray_img,10)
figure(1)
io.imshow(img_g2)
img_g3=filters.gaussian_filter(gray_img,5)
figure(2)
io.imshow(img_g3)
img_g4=filters.gaussian_filter(gray_img,2)
figure(3)
io.imshow(img_g4)
```

```
#Sobel edge detection
edge=filters.sobel(gray_img)
io.imshow(edge)
```

```
#Canny edge detection
from skimage import feature
edge_canny=feature.canny(gray_img,3)
io.imshow(edge_canny)
```
